1、有 50 个人围成一个圈，按顺时针报数。从第一个人开始报数（从 1 到 3 报数），凡报到 3 的人退出圈子，问最后留下的是原来多少号的位置？


```python
def skip_by(l, skip=3, count=0):
    if len(l) == 1:
        return l
    new_l = l[:]
    for i, _ in enumerate(new_l):
        count += 1
        if (count % skip == 0):
            l.remove(new_l[i])
    return skip_by(l, skip, count)


def skip_by2(l, skip=3, count=1):
    ln = len(l)
    if ln == 1:
        return l
    return skip_by2(
        [
            v
            for i, v
            in enumerate(l)
            if (count + i) % skip != 0
        ],
        skip,
        count + ln)


n = 50
li = [i for i in range(1, n + 1)]
print(skip_by(li))
print(skip_by2(li))
```

3、单词统计，输入以下英文短文，请用 Python 编写一段程序，实现每个单词出现在次数。范文如下：
There are moments in life when you miss someone so much that you just want to pick them from your dreams and hug them for real! Dream what you want to dream;go where you want to go;be what you want to be,because you have only one life and one chance to do all the things you want to do.
　　 May you have enough happiness to make you sweet,enough trials to make you strong,enough sorrow to keep you human,enough hope to make you happy? Always put yourself in others’shoes.If you feel that it hurts you,it probably hurts the other person, too.


```python
def split_str(t):
    return [c.lower() for line in t.split('\n')
            for c in line.split(' ') if c.strip() != '']


def count_word_by_group(l):
    return [
        (k, len(list(g)))
        for k, g in itertools.groupby(sorted(l))
    ]


sorted_by_count = functools.partial(
    sorted,
    key=lambda item: item[1],
    reverse=True
)


def word_count(s):
    lst = split_str(s)
    counted = count_word_by_group(lst)
    return sorted_by_count(counted)


def handle_dict(d, key):
    this_d = d.copy()
    this_d[key] = this_d.get(key, 1) + 1
    return this_d


def word_count2(s):
    lst = split_str(s)
    d = functools.reduce(handle_dict, lst, dict())
    tuple_list = [(k, c) for k, c in d.items()]
    return sorted_by_count(tuple_list)


t = """There are moments in life when you miss someone so much that you just want to pick them from your dreams and hug them for real! Dream what you want to dream;go where you want to go;be what you want to be,because you have only one life and one chance to do all the things you want to do.

May you have enough happiness to make you sweet,enough trials to make you strong,enough sorrow to keep you human,enough hope to make you happy? Always put yourself in others’shoes.If you feel that it hurts you,it probably hurts the other person, too."""

out1 = word_count(t)
for k, c in out1:
    print(f'{k}: {c}')

out2 = word_count(t)
for k, c in out2:
    print(f'{k}: {c}')
```

4、场景：利用程序实现个人所得税计算器。

具体计算示例：

某职员每月应发工资均为 30000 元，每月减除个人起征点费用 5000 元，公积金
为 2500 元，社保费用为 2000 元，专项附加费为 2000 元。

以前 3 个月为例，应当按照以下方法计算各月应预扣预缴税额：

- 1 月份：$(30000 - 5000 - 2500 - 2000 - 2000)\times3\%=555$ 元；
- 2 月份：$(30000\times2 - 5000\times2 - 2500\times2 - 2000\times2 - 2000\times2)\times10\% - 2520 - 555=625$ 元；
- 3 月份：$(30000\times3 - 5000\times3 - 2500\times3 - 2000\times3 - 2000\times3)\times10\% - 2520 - 555 - 625=1850$ 元

个人所得税预扣率表一

| 级数 | 累积预扣预缴应纳税所得额         | 预扣率(%) | 速算扣除数 |
| :--: | -------------------------------- | --------: | ---------: |
|  1   | 不超过 36000 元                  |         3 |          0 |
|  2   | 超过 36000 元至 144000 元的部分  |        10 |       2520 |
|  3   | 超过 144000 元至 300000 元的部分 |        20 |      16920 |
|  4   | 超过 300000 元至 420000 元的部分 |        25 |      31920 |
|  5   | 超过 420000 元至 660000 元的部分 |        30 |      52920 |
|  6   | 超过 660000 元至 960000 元的部分 |        35 |      85920 |
|  7   | 超过 960000 元的部分             |        45 |     181920 |


```python
import math

# personal income tax list
income_tax_list = [
    (-math.inf, 36000, 0.03, 0),
    (36000, 144000, 0.1, 2520),
    (144000, 300000, 0.2, 16920),
    (300000, 420000, 0.25, 31920),
    (420000, 660000, 0.3, 52920),
    (660000, 960000, 0.35, 85920),
    (960000, math.inf, 0.45, 181920),
]
# monthly income
income_month = 30000
# monthly deduction
deduction_month = 5000 + 2500 + 2000 + 2000
# (monthly income, accumulative income, deduction, taxable-salary)
fee_list = [
    (
        income_month,
        i * income_month,
        i * deduction_month,
        i * income_month - i * deduction_month
    )
    for i
    in range(1, 13)
]


def get_taxrate_quick_deduction(taxable_salary, income_tax_list):
    """get taxrate and quick deduction by taxable salary"""
    return next((taxrate, deduction)
                for s, e, taxrate, deduction in income_tax_list if s < taxable_salary <= e)


def calculate_tax_by_month(month, taxrate_list):
    """calculate tax for month"""
    return functools.reduce(
        lambda acc, el: acc + [el[3] * el[4] - el[5] - sum(acc)],
        taxrate_list[:month],
        [0]
    )[1:]


# (monthly income, accumulative income, deduction, taxable-salary, taxrate, quick-deduction)
taxrate_list = [
    (
        monthly_income,
        income,
        deduction,
        actual,
        get_taxrate_quick_deduction(actual, income_tax_list)[0],
        get_taxrate_quick_deduction(actual, income_tax_list)[1]
    )
    for monthly_income, income, deduction, actual
    in fee_list
]
# tax
tax_list = calculate_tax_by_month(12, taxrate_list)
# (monthly income, accumulative income, deduction, taxable-salary, taxrate, quick-deduction, tax)
fee_tax_list = [
    (*taxrate, tax)
    for taxrate, tax
    in zip(taxrate_list, calculate_tax_by_month(12, taxrate_list))
]
for idx, el in enumerate(fee_tax_list):
    print(f'{" " + str(idx+1) + "month ":-^50}')
    print(
        f'{"salary income:":>30}' \
        f'{el[0]:>12,.2f}'
    )
    print(
        f'{"accumulative salary income:":>30}' \
        f'{el[1]:>12,.2f}'
    )
    print(
        f'{"accumulative deduction:":>30}' \
        f'{el[2]:>12,.2f}'
    )
    print(
        f'{"taxable salary:":>30}' \
        f'{el[3]:>12,.2f}'
    )
    print(
        f'{"taxrate:":>30}' \
        f'{el[4]:>12,.2f}'
    )
    print(
        f'{"quick deduction:":>30}' \
        f'{el[5]:>12,.2f}'
    )
    print(
        f'{"tax:":>30}' \
        f'{el[6]:>12,.2f}'
    )
    print('-'*50+'\n')
print(
    f'total salary of year: {sum([el[0] for el in fee_tax_list]):,.2f}\n\
total tax of year: {sum([el[6] for el in fee_tax_list]):,.2f}')
```

```python
# 实现个人所得税计算
income_list = list()
for i in range(1, 13):
    month1 = 30000
    income_list.append(month1)
# 公积金
gjj = 2500
# 社保
shebao = 2000
# 专项附加
zhuanxiang = 2000
# 总扣税
total_tax = 0
# 总收入
total_income = 0
# 计算累积的总扣除项


def compute_month_deduction(month, gjj, shebao, zhuanxiang):
    final_deduct = (gjj+shebao+zhuanxiang+5000)*(month)
    # print(f"第{month}月总扣除项金额={final_deduct}元）")
    return final_deduct
# 计算每月应预扣预缴税额


def compute_tax(month, income, tax):
    if 0 <= income <= 36000:
        tax_money = income*0.03-tax
    elif 36000 <= income <= 144000:
        tax_money = income*0.1-2520-tax
    elif 144000 <= income <= 300000:
        tax_money = income*0.2-16920-tax
    elif 300000 <= income <= 420000:
        tax_money = income*0.25-31920-tax
    elif 420000 <= income <= 660000:
        tax_money = income*0.3-52920-tax
    elif 66000 <= income <= 960000:
        tax_money = income*0.35-85920-tax
    elif 960000 <= income:
        tax_money = income*0.45-181920-tax
    else:
        tax_money = 0
    print("{}月最终的扣税人民币为:{}元".format(month, tax_money))
    return tax_money


for month in range(1, 13):
    print('----')
    cur_income = int(income_list[month-1])
    print(f"{month}月收入为：{cur_income}元")
    total_income = cur_income+total_income
    last_tax = total_tax
    # 总扣除项
    final_deduct = compute_month_deduction(month, gjj, shebao, zhuanxiang)
    # 计税总额=总收入-总扣除项
    income = total_income-final_deduct
    cur_tax = compute_tax(month, income, last_tax)
    # print(f"本月扣税金额={cur_tax}")
    total_tax = last_tax+cur_tax

print("----")
print(f"年度总收入={total_income}元")
print(f"年度总扣税={total_tax}元")
```

6、用 Python 实现一个排序：
请定义一个函数 mySort，参数为一个列表，参数列表中的元素都是整数.
mySort 函数需要将参数列表中的元素按从小到大排序，最终返回一个新的 list。
请按下面算法的思路实现函数：

1. 创建一个新的列表 newList
2. 先找出所有元素中最小的，append 在 newList 里面
3. 再找出剩余的所有元素中最小的，append 在 newList 里面
4. 依次类推，直到所有的元素都放到 newList 里面


```python
import math
def mySort(in_list):
    newList = []
    while True:
        if len(in_list) <= 0:
            break
        minVal = math.inf
        for val in in_list:
            if val < minVal:
                minVal = val
        in_list.remove(minVal)
        newList.append(minVal)
    return newList

    
```

```python
myList = [1, 3, 5, 2, 9, 10, 0, 1]
print(mySort(myList))
```
